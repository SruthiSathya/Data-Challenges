```
import pandas as pd
fdata = pd.read_csv("cars.csv")

```

```
# Checking for duplicate columns 
fdata = fdata.drop_duplicates(keep='first')
fdata.duplicated().sum()
```

```
'''
Chcek for null values in the dataset 
'''
fdata.shape
fdata.isnull().sum()
```

```
import category_encoders as ce 
from sklearn.model_selection import train_test_split

cat_cols = ["Model","Fuel_Type"]
oe = ce.OrdinalEncoder(cols = cat_cols)
fdata = oe.fit_transform(fdata)
#fdata.head()
ncols = fdata.shape[1]
#spilitting into x and y
x = fdata.iloc[:,0:ncols-1]
y = fdata.iloc[:,ncols-1:ncols]
train_x,test_x,train_y,test_y = train_test_split(x, y, test_size = 0.20, random_state = 0)

print(ncols)
```
